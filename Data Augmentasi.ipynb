# Import Library

```python
import numpy as np
from numpy import expand_dims
import matplotlib.pyplot as plt
import tensorflow as tf
from keras.preprocessing.image import load_img, img_to_array
from keras.preprocessing.image import ImageDataGenerator
```

```python
# menampilkan horizontal flip
image = load_img('/content/drive/MyDrive/Data Augmentasi/train/jahe/jahe (80).jpg')
dataImage = img_to_array(image)
imageNew = expand_dims(dataImage, 0)

horizontalflip = ImageDataGenerator(horizontal_flip=True)
iterator = horizontalflip.flow(imageNew, batch_size=32)
plt.figure(figsize=(10, 10))
for i in range(9):
	plt.subplot(3,3, 1 + i, 
	            # title='horizontalflip'
              )
	batch = iterator.next()
	image = batch[0].astype('uint8')
	plt.imshow(image)
plt.show()
```
